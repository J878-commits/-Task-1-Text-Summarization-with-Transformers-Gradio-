```python
# 📱 Instagram UI/UX Redesign: Navigation Clarity
# Author: Josy | CodTech Internship Task-3

from IPython.display import Markdown, display, Image

def show_section(title, content):
    display(Markdown(f"## {title}\n{content}"))

# 🔍 Pain Point
show_section("🔍 Pain Point", """
Instagram users often struggle to differentiate between Reels, Stories, and Feed due to:
- Overlapping swipe gestures
- Similar icons and transitions
- Lack of clear labels or visual boundaries

This leads to accidental exits, missed content, and cognitive overload.
""")

# 🎯 Redesign Goals
show_section("🎯 Redesign Goals", """
- Clarify content types with distinct visual cues
- Streamline navigation using intuitive gestures and labels
- Enhance discoverability without overwhelming users
- Improve accessibility across age groups and devices
""")
```

```python
import pandas as pd

# 📊 Before vs After Comparison
data = {
    "Feature": [
        "Swipe Gestures",
        "Content Labels",
        "Icon Design",
        "User Feedback",
        "Accessibility"
    ],
    "Before Redesign": [
        "Overlapping gestures for Reels, Stories, Feed",
        "Minimal or no labels; relies on user memory",
        "Similar icons across modes",
        "Frequent confusion, accidental exits",
        "Limited support for older users or low-vision"
    ],
    "After Redesign": [
        "Distinct gestures per content type (e.g., vertical for Reels, horizontal for Stories)",
        "Clear, persistent labels with subtle animations",
        "Color-coded and shape-differentiated icons",
        "Improved clarity, fewer navigation errors",
        "High-contrast mode, larger tap targets"
    ]
}

df = pd.DataFrame(data)
display(Markdown("## 📊 Before vs After Comparison"))
display(df)
```

```python
from graphviz import Digraph

# 🧭 Redesigned Navigation Flow
dot = Digraph(comment='Instagram Navigation Flow')

dot.attr(rankdir='LR', size='8,5')

# Nodes
dot.node('F', '📸 Feed')
dot.node('S', '🕒 Stories')
dot.node('R', '🎬 Reels')
dot.node('E', '🔍 Explore')
dot.node('P', '👤 Profile')

# Edges
dot.edge('F', 'S', label='Swipe Right')
dot.edge('S', 'R', label='Swipe Down')
dot.edge('R', 'F', label='Swipe Up')
dot.edge('F', 'E', label='Tap Explore Icon')
dot.edge('F', 'P', label='Tap Profile Icon')

# Render
display(Markdown("## 🧭 Redesigned Navigation Flow"))
dot.render('nav_flow', format='png', cleanup=True)
Image(filename='nav_flow.png')
```

```python
import matplotlib.pyplot as plt

# 📈 Simulated User Feedback Scores
labels = ['Navigation Clarity', 'Content Recognition', 'Gesture Intuitiveness', 'Visual Appeal', 'Accessibility']
before = [2.5, 2.8, 2.2, 3.0, 2.4]
after = [4.5, 4.6, 4.3, 4.7, 4.5]

x = range(len(labels))

plt.figure(figsize=(10, 6))
plt.bar(x, before, width=0.4, label='Before', color='salmon', align='center')
plt.bar([i + 0.4 for i in x], after, width=0.4, label='After', color='mediumseagreen', align='center')
plt.xticks([i + 0.2 for i in x], labels, rotation=30)
plt.ylabel('User Rating (out of 5)')
plt.title('📈 User Feedback: Before vs After Redesign')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
