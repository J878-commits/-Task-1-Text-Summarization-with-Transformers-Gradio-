‚úÖ Option 1: Load Built-in Dataset (No Download Needed)

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target, name='species')
```

üß™ Step 2: Split the Data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

üå≥ Step 3: Train the Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X_train, y_train)
```

üìä Step 4: Evaluate the Model

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = model.predict(X_test)

print("‚úÖ Accuracy:", accuracy_score(y_test, y_pred))
print("\nüìä Classification Report:\n", classification_report(y_test, target_names=iris.target_names))
print("\nüîç Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
print("‚úÖ Accuracy:", accuracy_score(y_test, y_pred))
print("\nüìä Classification Report:\n", classification_report(y_test, y_pred, target_names=iris.target_names))
print("\nüîç Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

üé® Step 5: Visualize the Decision Tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plot_tree(
    model,
    feature_names=X.columns,
    class_names=iris.target_names,
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("üå≥ Decision Tree for Iris Classification", fontsize=14)
plt.show()
```

üè∑Ô∏è Understanding Decision Tree Labels
Each node in the Decision Tree visualization contains key metrics that explain how the model makes decisions:

True/False Branches These represent the outcome of the split condition at each node. ‚Üí If the condition (e.g., petal length (cm) ‚â§ 2.45) is true, follow the left branch. ‚Üí If false, follow the right branch.

gini Gini impurity measures how mixed the classes are at a node. ‚Üí A Gini of 0 means perfect purity (all samples belong to one class). ‚Üí The algorithm aims to minimize Gini at each split.

samples The number of training samples that reach that node. ‚Üí Helps track how data is distributed across the tree.

value The class distribution at that node. ‚Üí Example: [35, 35, 35] at the root means equal representation of all three Iris species. ‚Üí As you move down, these values reflect how samples are split by feature thresholds.

class The predicted class for samples reaching that node. ‚Üí At leaf nodes, this is the final prediction for any sample that follows the path to that leaf.

üå≥ Summary
The tree begins with all data at the root and recursively splits based on feature thresholds to separate classes. Each split aims to reduce impurity and increase class purity, leading to confident predictions at the leaf nodes.
