```python
solidity_code = """
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract CodTechToken {
    string public name = "CodTechToken";
    string public symbol = "CTT";
    uint8 public decimals = 18;
    uint256 public totalSupply;

    mapping(address => uint256) public balanceOf;
    mapping(address => mapping(address => uint256)) public allowance;

    address public owner;

    event Transfer(address indexed from, address indexed to, uint256 value);
    event Approval(address indexed owner, address indexed spender, uint256 value);

    constructor(uint256 _initialSupply) {
        owner = msg.sender;
        totalSupply = _initialSupply * (10 ** uint256(decimals));
        balanceOf[owner] = totalSupply;
        emit Transfer(address(0), owner, totalSupply);
    }

    function transfer(address _to, uint256 _value) public returns (bool success) {
        require(balanceOf[msg.sender] >= _value, "Insufficient balance.");
        balanceOf[msg.sender] -= _value;
        balanceOf[_to] += _value;
        emit Transfer(msg.sender, _to, _value);
        return true;
    }

    function approve(address _spender, uint256 _value) public returns (bool success) {
        allowance[msg.sender][_spender] = _value;
        emit Approval(msg.sender, _spender, _value);
        return true;
    }

    function transferFrom(address _from, address _to, uint256 _value) public returns (bool success) {
        require(balanceOf[_from] >= _value, "Insufficient balance.");
        require(allowance[_from][msg.sender] >= _value, "Allowance exceeded.");
        balanceOf[_from] -= _value;
        balanceOf[_to] += _value;
        allowance[_from][msg.sender] -= _value;
        emit Transfer(_from, _to, _value);
        return true;
    }
}
"""
print(solidity_code)
```

## üîÑ Step 3: Token Transfer Simulation

Once the CodTechToken contract is deployed, we can interact with it using functions like:

- `transfer(address _to, uint256 _value)`
- `approve(address _spender, uint256 _value)`
- `transferFrom(address _from, address _to, uint256 _value)`

Below is a Python simulation that mimics how balances and allowances change during these interactions.


```python
# Simulating token transfer logic in Python

class CodTechTokenSim:
    def __init__(self, initial_supply, owner):
        self.decimals = 18
        self.total_supply = initial_supply * (10 ** self.decimals)
        self.balance = {owner: self.total_supply}
        self.allowance = {}
        self.owner = owner

    def transfer(self, sender, recipient, amount):
        if self.balance.get(sender, 0) >= amount:
            self.balance[sender] -= amount
            self.balance[recipient] = self.balance.get(recipient, 0) + amount
            return f"‚úÖ Transfer of {amount} from {sender} to {recipient} successful."
        else:
            return "‚ùå Transfer failed: Insufficient balance."

    def approve(self, owner, spender, amount):
        if owner not in self.allowance:
            self.allowance[owner] = {}
        self.allowance[owner][spender] = amount
        return f"‚úÖ {spender} approved to spend {amount} on behalf of {owner}."

    def transfer_from(self, spender, from_addr, to_addr, amount):
        if self.balance.get(from_addr, 0) >= amount and self.allowance.get(from_addr, {}).get(spender, 0) >= amount:
            self.balance[from_addr] -= amount
            self.balance[to_addr] = self.balance.get(to_addr, 0) + amount
            self.allowance[from_addr][spender] -= amount
            return f"‚úÖ {spender} transferred {amount} from {from_addr} to {to_addr}."
        else:
            return "‚ùå TransferFrom failed: Check balance or allowance."

# Example usage
token = CodTechTokenSim(initial_supply=1000, owner="Josy")
print(token.transfer("Josy", "Alice", 100))
print(token.approve("Alice", "Bob", 50))
print(token.transfer_from("Bob", "Alice", "Charlie", 30))
```
